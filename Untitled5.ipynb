```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a DataFrame
data = {
    'Condition': [
        'Simple vascular calcification score = 0',
        'Simple vascular calcification score > 0'
    ],
    'Number_of_patients': [84, 122]
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
# Set the style of the visualization
sns.set(style="whitegrid")

# Create a bar plot
plt.figure(figsize=(10, 6))  # Specify the size of the plot
sns.barplot(x='Condition', y='Number_of_patients', data=df, palette='muted')

# Add titles and labels
plt.title('Number of Patients by Vascular Calcification Score')
plt.xlabel('Vascular Calcification Score')
plt.ylabel('Number of Patients')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.tight_layout()  # Adjust the plot to fit into the figure area.
plt.show()
```
